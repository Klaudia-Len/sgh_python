5. BUDOWA MODELU

PODZIAL DANYCH NA ZBIOR UCZACY I ZBIOR TESTOWY 

```python
X = realestate[['X2 house age', 'X3 distance to the nearest MRT station', 'X4 number of convenience stores',
               'X5 latitude', 'X6 longitude']]
y = realestate['Y house price of unit area']
```

```python
pip install scikit-learn
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=40)
```

Zbiór uczący dla modelu zawiera 70% z wszystkich obserwacji, natomiast zbiór testowy 30% obserwacji. 

5.1 ESTYMACJA MODELU ZA POMOCĄ FUNKCJI SKLEARN

```python
from sklearn.linear_model import LinearRegression
```

```python
lm = LinearRegression()
```

```python
lm.fit(X_train,y_train)
```

```python
factor = pd.DataFrame(lm.coef_,X.columns,['Factors'])
print(factor)
print(lm.intercept_)
```

Model regresji liniowej utworzonej za pomocą funkcji sklearn ma postać:
Y house price of unit area= -0.297851 * X2 house age -0.004402 * X3 distance to the nearest MRT station  + 1.114714 * X4 number of convenience stores + 270.105450 * X5 latitude -8.762072 * X6 longitude -5635.5348

INTERPRETACJA PARAMETRÓW MODLEU
- przy zachowaniu stałego poziomu pozostałych parametrów, jednostkowa zmiana parametru 'X2 house age' spowoduje spadek zmiennej 'Y house price of unit area' o 0.297851 
- przy zachowaniu stałego poziomu pozostałych parametrów, jednostkowa zmiana parametru 'X3 distance to the nearest MRT station' spowoduje spadek zmiennej 'Y house price of unit area' o 0.004402 
- przy zachowaniu stałego poziomu pozostałych parametrów, jednostkowa zmiana parametru 'X4 number of convenience stores' spowoduje wzrost zmiennej 'Y house price of unit area' o 1.114714 
- przy zachowaniu stałego poziomu pozostałych parametrów, jednostkowa zmiana parametru ' X5 latitude' spowoduje wzrost zmiennej 'Y house price of unit area' o 270.105450
- przy zachowaniu stałego poziomu pozostałych parametrów, jednostkowa zmiana parametru ' X6 longitude' spowoduje spadek zmiennej 'Y house price of unit area' o 8.762072 

5.1.1. PREDYKCJA MODELU SKLEARN

```python
predictions = lm.predict(X_test)
plt.scatter(y_test,predictions)
```

```python
sns.distplot((y_test-predictions),bins=40);
```

5.2 ESTYMACJA MODELU ZA POMOCĄ FUNKCJI STATSMODELS

```python
import statsmodels.api as sm
 
```

```python
model = sm.OLS(y_train, X_train)
results = model.fit()
print(results.summary())
```

Model regresji liniowej utworzonej za pomocą funkcji statsmodels ma postać:
Y house price of unit area= -0.2986 * X2 house age -0.0049 * X3 distance to the nearest MRT station  + 1.1123 * X4 number of convenience stores + 260.2014 * X5 latitude   -53.0931 * X6 longitude

INTERPRETACJA PARAMETRÓW MODLEU
- przy zachowaniu stałego poziomu pozostałych parametrów, jednostkowa zmiana parametru 'X2 house age' spowoduje spadek zmiennej 'Y house price of unit area' o 0.2986 
- przy zachowaniu stałego poziomu pozostałych parametrów, jednostkowa zmiana parametru 'X3 distance to the nearest MRT station' spowoduje spadek zmiennej 'Y house price of unit area' o 0.0049 
- przy zachowaniu stałego poziomu pozostałych parametrów, jednostkowa zmiana parametru 'X4 number of convenience stores' spowoduje wzrost zmiennej 'Y house price of unit area' o 1.1123 
- przy zachowaniu stałego poziomu pozostałych parametrów, jednostkowa zmiana parametru ' X5 latitude' spowoduje wzrost zmiennej 'Y house price of unit area' o 260.2014
- przy zachowaniu stałego poziomu pozostałych parametrów, jednostkowa zmiana parametru ' X6 longitude' spowoduje spadek zmiennej 'Y house price of unit area' o  53.0931 

WŁASNOŚCI STATYSTYCZNE MODELU 
Według powyższego modelu, przy założeniu poziomu istotności 5%, wszystkie zmienne objaśniające są statystycznie istotne (wartość p-value mniejsza niż 0,05). Wartość statystyki R^2 jest bliska 1, model dobrze wyjaśnia wartości rzeczywiste. Test  Jarque-Bera wskazuje że reszty modelu pochodzą z rozkładu normalnego. 

5.2.1. ESTYMACJA PARAMETRÓW MODELU STATSMODELS

```python
predictionsStats = results.predict(X_test)
plt.scatter(y_test,predictionsStats)
```

6. OCENA DOPASOWANIA MODELU NA ZBIORZE TESTOWYM 

```python
from sklearn import metrics
```

6.1 Ocena modelu sklearn

```python
print('MAE:', metrics.mean_absolute_error(y_test, predictions))
print('MSE:', metrics.mean_squared_error(y_test, predictions))
print('RMSE:', np.sqrt(metrics.mean_squared_error(y_test, predictions)))
```

Wartości rzeczywiste, przeciętnie odchylają się o 5,79 od wartości estymowanych przez modelu. Model w trakcie estymacji, przeciętnie odchyla się o 7,61 od wartości rzeczywistej. Pomiedzy wartością MAE oraz RMSE różnica jest nieznaczna, zatem estymowane wartości nie odbiegają znacznie od wartości rzeczywistych.

6.2 Ocena modelu statsmodel

```python
print('MAE:', metrics.mean_absolute_error(y_test, predictionsStats))
print('MSE:', metrics.mean_squared_error(y_test, predictionsStats))
print('RMSE:', np.sqrt(metrics.mean_squared_error(y_test, predictionsStats)))
```

Wartości rzeczywiste, przeciętnie odchylają się o 5,75 od wartości estymowanych przez modelu. Model w trakcie estymacji, przeciętnie odchyla się o 7,61 od wartości rzeczywistej. Pomiedzy wartością MAE oraz RMSE różnica jest nieznaczna, zatem estymowane wartości nie odbiegają znacznie od wartości rzeczywistych.

Modele utworzone za pomocą obu funkcji nie różnią się znacząco od siebie. Błędy dla obydwu modeli są podobne.
